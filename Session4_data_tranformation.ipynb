```python
math_score = df['math'].to_numpy()
print(math_score)
print(math_score.mean())
```

```python
science_score = df['science'].to_numpy()
print(science_score)
```

```python
eng_score = df['english'].to_numpy()
print(eng_score)
```

```python
final_project_score = df['final_project_score'].to_numpy()
print(final_project_score)
```

```python
avg_score=(math_score+science_score+eng_score+final_project_score)/4
print(avg_score)
```

```python
df = pd.read_csv('session4_dummy_students_data.csv')
df['avg_score'] = avg_score
print(df.head())
```

```python
top_students = df[df['avg_score']>85]
print(top_students)
```

```python
low_attendance = df[df['attendance_percent'] < 75]
print(low_attendance)
```

```python
student_city = df[df['city'] == "Mathura"]
print(student_city)
```

```python
less_active = df[df['assignments_submitted']<6]
print(less_active)
```

```python
df.to_csv('session4_dummy_students_data.csv',index = False)
```
